# Language Translator

## Preprocessor

Basially removes the comments, etc... To only have the main code required, that is converts source code to pure HLL

## Compiler

Converts HLL to Assembly Language

## Asssembler

Converts Assembly language "Relocatable" Machine Code

What does relocatable mean? 

When we run a program, only then is it converted to a process and is transferred to main memory, so before eecution, the sequene of 1's and 0's are given a reallocatable memory, for example, i+0, i+1, i+2, so on

## Linker/Loader

Converts the relocatable machine code to absolute machine code, for example, here, the i+0, i+1, so on is converted to absolute addresses
