```python
grammar = r"""
grammar PythonSubset;

program : statement+ ;

statement : assignment ;

assignment : ID '=' NUMBER;

ID : [a-zA-Z_][a-zA-Z_0-9]* ;
NUMBER : [0-9]+ ;
WS : [ \t\r\n]+ -> skip ;
"""

with open("PythonSubset.g4","w") as f:
    f.write(grammar)
```

```python
!java -jar antlr-4.13.1-complete.jar -Dlanguage=Python3 PythonSubset.g4
```

```python
from antlr4 import *
from PythonSubsetLexer import PythonSubsetLexer
from PythonSubsetParser import PythonSubsetParser


code = "x = 5"

input_stream = InputStream(code)
lexer = PythonSubsetLexer(input_stream)
stream = CommonTokenStream(lexer)
parser = PythonSubsetParser(stream)

tree = parser.program()

print(tree.toStringTree(recog=parser))
```

```python
!java -jar antlr-4.13.1-complete.jar -Dlanguage=Python3 -visitor PythonSubset.g4 
```

The above line of code runs a terminal command in jupyter(that's why we have a '!' at the beginning and creates a visitor for going through and reading the parse tree created above

```python
#Now we change the above visitAssignment method to return java assignment statement
from PythonSubsetVisitor import PythonSubsetVisitor

class MyVisitor(PythonSubsetVisitor):

    def visitAssignment(self, ctx):
        var_name = ctx.ID().getText()
        value = ctx.NUMBER().getText()

        return f"int {var_name} = {value};"
```

```python
visitor = MyVisitor()
print(visitor.visit(tree))
```
